```python
import tensorflow as tf
import numpy as np
import matplotlib as mp
import pandas as pd
from sklearn.model_selection import train_test_split # very handy way of selecting sets
```

```python
df=pd.read_csv('Dataset/Breast_Cancer.csv ')
```

```python
df.head()
```

```python
df = df.drop(['Unnamed: 32'],axis=1)
```

```python
df.head()
```

```python
df.info()
```

```python
def change(a):
    if a=="M":
        return 1
    else:
        return 0
```

```python
df["diagnosis"]=df['diagnosis'].apply(change)
```

```python
df['diagnosis'].value_counts()
```

```python
labels_df = df['diagnosis']
labels_df.head()
labels_df.shape
```

```python
labels_df = pd.get_dummies(labels_df)
labels_df.shape
```

```python
features_df=df.drop(['id'],axis=1)
features_df=features_df.drop(['diagnosis'],axis=1)
```

```python
print(features_df.shape)
print(labels_df.shape)
```

```python
# Train Test Split
# Training Data = 80% of Dataset
# Test Data = 20% of Dataset
X_train, X_test, y_train, y_test = train_test_split(features_df, labels_df, test_size=0.2, random_state=1, stratify=labels_df)
```

```python
X_train
```

```python
y_train
```

```python
y_train=np.array(y_train)
y_test=np.array(y_test)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
m,n=X_train.shape
print(m)
print(n)
```

```python
y_train.shape
```

```python
#Tensorflow Start
```

```python
learning_rate = 0.01
training_epochs = 500

tf.reset_default_graph()

# By aving 2 features: hours slept & hours studied
X = tf.placeholder(tf.float32, [None, n], name="X")
Y = tf.placeholder(tf.float32, [None, 2], name="Y")
```

```python
# Initialize our weigts & bias
W = tf.get_variable("W", [n, 2], initializer = tf.contrib.layers.xavier_initializer())
b = tf.get_variable("b", [2], initializer = tf.zeros_initializer())
print(W.shape)
```

```python
Z = tf.add(tf.matmul(X, W), b)
prediction = tf.nn.sigmoid(Z)
```

```python
cost = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits = Z, labels = Y))
```

```python
optimizer = tf.train.AdamOptimizer(learning_rate).minimize(cost)
```

```python
init = tf.global_variables_initializer()

cost_history = np.empty(shape=[1],dtype=float)
```

```python
with tf.Session() as sess:
    sess.run(init)
    
    for epoch in range(training_epochs):
        _, c = sess.run([optimizer, cost], feed_dict={X:X_train, Y: y_train})
        print("Epoch:", '%04d' % (epoch+1), "cost=", "{:.9f}".format(c), \
               "W=", sess.run(W), "b=", sess.run(b))
        cost_history = np.append(cost_history, c)
        
        
    # Calculate the correct predictions
    correct_prediction = tf.to_float(tf.greater(prediction, 0.5))

    # Calculate accuracy on the test set
    accuracy = tf.reduce_mean(tf.to_float(tf.equal(Y, correct_prediction)))
    print ("Accuracy:", accuracy.eval({X:X_train, Y:y_train}))
    print ("test Accuracy:", accuracy.eval({X:X_test, Y:y_test}))
    
```
